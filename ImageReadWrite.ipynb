<h2>Image Reading</h2> 

cv2.imread(filename, flag)

Parameters:<p>
filename: A string representing the file name. The filename must include image format like .jpg, .png, etc.
<p>flag: it is given to decide mode of file reading 
        RGB,GRAY or RGBA. <p> RGB -> cv2.IMREAD_COLOR or 1<p> GRAY -> cv2.IMREAD_GRAYSCALE 0 <p> RGBA -> cv2.IMREAD_UNCHANGED or -1 <p> This function returns numpy array so any operaion on array we can apply.
Return Value: It returns numpy array if read successfully otherwise will retrun None.


```python
img = cv2.imread("./images/dog.jpg",cv2.IMREAD_COLOR) 
```

once image is read from disk then we can display. But jupyter notebook does not work on opencv imshow function. Here we will use matplotlib

```python
from matplotlib import pyplot as plt
plt.imshow(img)
plt.show()
```

if you see the output not looks like origianl image reason is opencv read data in BGR format and matplot assumes its RGB. This we can convert using opencv function from BGR to RGB

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt
plt.imshow(img)
plt.show()
```

<h1>Image Writing </h1>

cv2.imwrite(filename, image)

Parameters:
filename: A string representing the file name. The filename must include image format like .jpg, .png, etc.
image: It is the image that is to be saved.

Return Value: It returns true if image is saved successfully.

Note : Before writing image to disk please again convert data in BGR becasue opencv understands this only.

```python
status = cv2.imwrite("test.jpg",img)
```
